PREPROCESSING SETUP:

From the small EDA that we have donr, we noticed that there is a lot of outliers in the Amount coulumn, and in the preprocessing part, we cant use standard scaler on it since standard scaler is heavily impacted by outliers. So, we are going to use RobustScaler which is resistant to extreme values. (same with Time column)

```python
from sklearn.model_selection import train_test_split
X= data.drop(columns="Class")
y= data["Class"]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,stratify = y)
```

```python
from sklearn.preprocessing import RobustScaler
amount_scaler = RobustScaler()
time_scaler = RobustScaler()
X_train['scaled_amount'] = amount_scaler.fit_transform(X_train[['Amount']])
X_train['scaled_time'] = time_scaler.fit_transform(X_train[['Time']])
X_test['scaled_amount'] = amount_scaler.transform(X_test[['Amount']])
X_test['scaled_time'] = time_scaler.transform(X_test[['Time']])
```

```python
X_train = X_train.drop(columns=['Amount','Time'])
X_test = X_test.drop(columns=['Amount','Time'])
```

```python
X_train.describe()
```

```python
X_train_scaled = X_train.copy()  
X_test_scaled = X_test.copy()
y_train = y_train.copy()
y_test = y_test.copy()
train_scaled = X_train_scaled.copy()
train_scaled['Class'] = y_train
test_scaled = X_test_scaled.copy()
test_scaled['Class'] = y_test
train_scaled.to_csv("creditcard_train_scaled.csv", index=False)
test_scaled.to_csv("creditcard_test_scaled.csv", index=False)
```
