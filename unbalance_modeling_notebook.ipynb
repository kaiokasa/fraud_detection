```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import os 
os.getcwd()
```

```python
train_data = pd.read_csv("creditcard_train_scaled.csv")
test_data = pd.read_csv("creditcard_test_scaled.csv")
train_data
```

```python
X_train = train_data.drop(columns="Class")
y_train = train_data["Class"]
print(X_train.shape,y_train.shape)
```

```python
X_test = test_data.drop(columns="Class")
y_test = test_data["Class"]
print(X_test.shape,y_test.shape)
```

Model training:

Lets create a function that will evaluate and store the models results

```python
import pandas as pd

results_df = pd.DataFrame(columns=[
    "Model",
    "Precision (Non-Fraud)", "Recall (Non-Fraud)", "F1 (Non-Fraud)",
    "Precision (Fraud)", "Recall (Fraud)", "F1 (Fraud)",
    "ROC-AUC"
])
```

Logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression(random_state=42)
```

```python
from crossval_report import print_crossval_results
lr_model = LogisticRegression(random_state=42)
crossval_result = print_crossval_results(lr_model, X_train, y_train, cv=5)
crossval_result
```

```python
lr_model.fit(X_train, y_train)
```

```python
y_test_pred = lr_model.predict(X_test)
y_test_proba = lr_model.predict_proba(X_test)[:, 1]
from store_results import store_and_print_results
results_df = store_and_print_results(results_df, "Logistic Rression (Unbalanced)", y_test, y_test_pred, y_test_proba)
results_df
```

Logistic Regression Results

1. Cross-Validation (on training set with CV = 5):

    Class 0 (Non-Fraud): Precision = 0.999, Recall = 1.000, F1 = 1.000

    Class 1 (Fraud): Precision = 0.871, Recall = 0.602, F1 = 0.712

    The model detects almost all non-fraud transactions perfectly.

    Fraud detection is weaker (recall 0.602), which is common with highly imbalanced datasets.

2. Test Set Evaluation:

    Class 0 (Non-Fraud): Precision = 0.999, Recall = 1.000, F1 = 1.000

    Class 1 (Fraud): Precision = 0.887, Recall = 0.643, F1 = 0.746

    Performance slightly improves on the test set for fraud detection.

1.  Cross-Validation (on training set)

        Class 0 (Non-Fraud): Precision, Recall, F1 = 1.0 → almost perfect at identifying legitimate transactions.

        Class 1 (Fraud):

        Precision = 0.945 → most predicted frauds are truly fraud.

        Recall = 0.787 → ~79% of actual frauds are correctly detected.

        F1 = 0.859 → solid balance between precision and recall for fraud.

2.  Test Set (Unbalanced Random Forest)

        Class 0 (Non-Fraud): Still perfect.

        Class 1 (Fraud):

        Precision = 0.950 → slightly better than CV.

        Recall = 0.776 → slightly lower than CV.

        F1 = 0.854 → very close to CV, showing good generalization.

Non-Fraud Class: Model is enar-perfetct for detecting non-Fraud Class

Fraud Class: Random Forest has higher precision (0.95 vs 0.887) than Logistic Regression, meaning fewer false positives. Recall is also higher (0.776 vs 0.643), meaning more frauds are correctly detected. F1 reflects this improvement.

During 5-fold cross-validation on the training set:

        Non-Fraud (Class 0): Precision, Recall, and F1 are essentially perfect (~1.000), meaning the model almost never misclassifies normal transactions.

        Fraud (Class 1): Precision=0.842, Recall=0.556, F1=0.670 — the model identifies many frauds correctly, but still misses a significant portion (recall is moderate).

On the final test set:

        Non-Fraud: Same excellent metrics as CV, showing stable performance.

        Fraud: Precision improved to 0.881, recall to 0.602, F1 to 0.715 — better than CV, likely due to having seen the full training set during final training.

SVM

```python
from sklearn.svm import SVC

svm_model = SVC(
    kernel='rbf',      
    C=1.0,             
    gamma='scale',
    probability=True,  
    random_state=42
)
from crossval_report import print_crossval_results
crossval_result = print_crossval_results(svm_model, X_train, y_train, cv=3)
crossval_result
```

```python
svm_model.fit(X_train,y_train)
```

```python
y_test_pred = svm_model.predict(X_test)
y_test_proba = svm_model.predict_proba(X_test)[:, 1]
from store_results import store_and_print_results
results_df = store_and_print_results(results_df, "SVM (Unbalanced)", y_test, y_test_pred, y_test_proba)
results_df
```

During 5-fold cross-validation, the SVM model achieved almost perfect performance for the non-fraud class, with precision, recall, and F1-score all close to 1.000. For the fraud class, the model obtained a precision of 0.937, recall of 0.645, and F1-score of 0.764. This indicates that when the model predicts fraud it is usually correct, but it still misses a noticeable number of fraudulent transactions.

When evaluated on the test set, the performance improved slightly. The fraud class achieved a precision of 0.971, recall of 0.684, and F1-score of 0.802. The ROC-AUC score of 0.940 indicates that the model has a strong ability to distinguish between fraudulent and legitimate transactions overall.
